```python
import pandas as pd
from google_play_scraper import Sort, reviews_all
import random
from sklearn.utils import shuffle
```

### Scraping reviews for fraud_apps

```python
fraud_apps=pd.read_json("fraud-apps.json")
fraud_app_revs=[]
for i,app in enumerate(fraud_apps["appId"]):
    result1=reviews_all(
    app,
    sleep_milliseconds=0,
    lang="en",
    country="us",
    sort=Sort.MOST_RELEVANT,
    filter_score_with=random.randint(3,5)
)
    result2=reviews_all(
    app,
    sleep_milliseconds=0,
    lang="en",
    country="us",
    sort=Sort.MOST_RELEVANT,
    filter_score_with=random.randint(1,3)
)
    
    
    if not result1 and not result2:
        continue #dont include the app, if there are no reviews
    
    
    result = [result1[i]["content"] for i in range(min(2, len(result1)))] + \
         [result2[i]["content"] for i in range(min(2, len(result2)))]
    fraud_app_revs.append([app, " ".join(result)])



fraud_revs_df=pd.DataFrame(fraud_app_revs,columns=["appId","reviews"])
    
```

```python
import pandas as pd
import random
import signal
from google_play_scraper import Sort, reviews_all

# Load the scraped apps CSV
scraped_apps = pd.read_csv("scraped_apps.csv")
scraped_app_revs2 = []

# Function to handle timeout
def timeout_handler(signum, frame):
    raise TimeoutError

# Set the timeout signal handler
signal.signal(signal.SIGALRM, timeout_handler)

# Loop through the apps, starting from index 80
for i, row in enumerate(scraped_apps[["appId", "Gemini_Label"]].iloc[80:].itertuples(index=False), start=80):
    print(i, row.appId, row.Gemini_Label)
    
    if i == 300:
        break  # Stop after processing 220 apps

    try:
        signal.alarm(13)  # Set timeout to 13 seconds

        # Fetch reviews
        result1 = reviews_all(
            row.appId,
            sleep_milliseconds=0,
            lang="en",
            country="us",
            sort=Sort.MOST_RELEVANT,
            filter_score_with=random.randint(3, 5)
        )

        result2 = reviews_all(
            row.appId,
            sleep_milliseconds=0,
            lang="en",
            country="us",
            sort=Sort.MOST_RELEVANT,
            filter_score_with=random.randint(1, 3)
        )

        signal.alarm(0)  # Cancel the alarm after success

    except TimeoutError:
        print(f"Skipping {row.appId} due to timeout.")
        continue  # Move to the next app if timeout occurs
    except Exception as e:
        print(f"Error processing {row.appId}: {e}")
        continue

    # Skip if both result1 and result2 are empty
    if not result1 and not result2:
        continue

    # Extract up to 2 reviews from each result
    result = [result1[i]["content"] for i in range(min(2, len(result1)))] + \
             [result2[i]["content"] for i in range(min(2, len(result2)))]

    scraped_app_revs2.append([row.appId, " ".join(result), row.Gemini_Label])

# Convert to DataFrame
```

```python
scraped_app_revs2_df=pd.DataFrame(scraped_app_revs2,columns=["appId","reviews","IsFraud"])
scraped_app_revs2_df.head()
```

## PRE-PROCESSING

```python
import pandas as
df_combined=pd.read_csv("app_reviews.csv")
```

```python
df_cleaned = df_combined.drop_duplicates()

print(df_cleaned.shape)
df_cleaned.to_csv("app_reviews.csv",index=False)
```

```python
df_combined=pd.concat([df_combined,scraped_app_revs2_df],ignore_index=True)
df_combined.shape
df_combined.to_csv("app_reviews.csv",index=False)
```

```python
df_cleaned
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df_=df_cleaned.copy(deep=True)
df_["IsFraud"]=le.fit_transform(df_["IsFraud"])
print(le.classes_)
df_.head()
```

```python
le.classes_
```

```python
import re
import string
def clean_text(text):
    text = text.lower()  # Lowercase
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

df_["reviews"] = df_["reviews"].apply(clean_text)
df_.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)  # Remove stopwords automatically
X = vectorizer.fit_transform(df_['reviews'])
```

```python
X.shape
```

```python
print(vectorizer.get_feature_names_out())  # Shows important words
print(X.toarray())  # TF-IDF transformed matrix
```

```python
Y=df_["IsFraud"]
assert X.shape[0] == Y.shape[0] , "error"
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
X_train.shape,y_test.shape
```

```python
Y.value_counts()
```

## ML Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
log_reg=LogisticRegression()

log_reg.fit(X_train,y_train)

y_pred_train=log_reg.predict(X_train)
y_pred_test=log_reg.predict(X_test)

print('train_acc',accuracy_score(y_train,y_pred_train))
print('test_acc',accuracy_score(y_test,y_pred_test))
print('train_rep',classification_report(y_train,y_pred_train))
print('test_rep',classification_report(y_test,y_pred_test))
print('train_cf',(confusion_matrix(y_train,y_pred_train)))
print('test_cf',(confusion_matrix(y_test,y_pred_test)))
```

	•	Precision: 0.00 (No correct predictions for Class 0)
	•	Recall: 0.00 (Model never identifies Class 0 correctly)
	•	F1-Score: 0.00
